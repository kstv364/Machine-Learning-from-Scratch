```python
import random as r,csv
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.rcParams['figure.figsize'] = [10, 10]
```

```python
import math
def phi(x):
    return 1/(1.0+math.exp(-x))

def dphi(x):
    return phi(x)*(1-phi(x))

def calcU1nj(n,j):
    s = 0
    for i in range(d+1):
        s += w1[j][i]*X[n][i]
    return s

def calcU2nk(n,k):
    s = 0
    for j in range(h+1):
        s += w2[k][j]*v1[n][j]
    return s

'''INITIALIZE THE WEIGHT MATRICES WITH RANDOM VALUES'''

h = 3 # nodes in hidden layer
C = 3 # no of classes
d = 2 # input dimensions


fixed_W1 = np.random.rand(h+1,d+1)
fixed_W2 = np.random.rand(C,h+1)
```

```python
sample_lengths = [500,1000,3000,5000,10000]
compare_results = {}
for N in sample_lengths:
    filename = 'non-linearly-separable-' + str(N) +'.csv'
    with open(filename, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["x1","x2","y1","y2","y3","class"])
        i = 0
        while(i<N):
            x = np.random.uniform(0,10)
            y = np.random.uniform(0,10)
            if((x-10)**2 + (y-10)**2 < 16):
                writer.writerow([str(x),str(y),"1","0","0","1"]); i+=1
            elif(x**2 + (y-10)**2 < 16):
                writer.writerow([str(x),str(y),"0","1","0","2"]); i+=1
            elif((x-5)**2 + y**2 < 16):
                writer.writerow([str(x),str(y),"0","0","1","3"]); i+=1
                
    df = pd.read_csv(filename)
    groups = df.groupby('class')
    fig, ax = plt.subplots()
    ax.margins(0.05) 
    for name, group in groups:
        ax.plot(group.x1, group.x2, marker='o', linestyle='', label = name)
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.rcParams['figure.figsize'] = [10, 10]
    plt.title('Dataset for size '+ str(N))
    plt.legend()
    plt.show()
    



    u1 = np.random.rand(N,h+1) #equivalent to u1[N][h+1]
    v1 = np.random.rand(N,h+1)
    u2 = np.random.rand(N,C)
    v2 = np.random.rand(N,C)

    w1 = np.copy(fixed_W1)
    w2 = np.copy(fixed_W2)
    
    df['x0'] = 1
    numberOfTrainingSamples = int(0.7*N)
    X = df[['x0','x1','x2']].values
    y = df[['y1','y2','y3','class']].values
    
    epoch = 200
    eta = 0.1
    errors = []
    error = 10**5 ; e = 0
    for e in range(epoch): # Convergence with presetting no of epochs
        error = 0
        for n in range(numberOfTrainingSamples): 
            for j in range(1,h+1):
                u1[n][j] = calcU1nj(n,j)
                v1[n][j] = phi(u1[n][j])
            err = 0
            for k in range(C):
                u2[n][k] = calcU2nk(n,k)
                v2[n][k] = phi(u2[n][k])
                err += (1/2)*((v2[n][k] - y[n][k])**2)

            error += err
            # Backprop
            for k in range(C):
                for j in range(h+1):
                    w2[k][j] += eta*(y[n][k]-v2[n][k])*dphi(u2[n][k])*v1[n][j]
            for i in range(d+1):
                for j in range(1,h+1):
                    w1[j][i] += eta*dphi(u1[n][j])*X[n][i]*sum([(y[n][k] - v2[n][k])*dphi(u2[n][k])*w2[k][j] for k in range(C)])
        error = error/numberOfTrainingSamples
        if (e)%10 ==0:
            print("Epoch :", e, "In-sample Error : ", error) 
        errors.append(error)
        # if(error<0.001):
        #     print("Breaking out since error is less than 0.001")
        #     break

    print('Final In-sample Error : ',errors[-1])
    
    '''TESTING PHASE'''
    testerrors = []
    numberOfTestSamples = N-numberOfTrainingSamples
    correct = 0
    for n in range(numberOfTrainingSamples,N):
        for j in range(1,h+1):
            u1[n][j] = calcU1nj(n,j)
            v1[n][j] = phi(u1[n][j])
        error = 0
        for k in range(C):
            u2[n][k] = calcU2nk(n,k)
            v2[n][k] = phi(u2[n][k])
            error += (1/2)*(v2[n][k] - y[n][k])**2

        testerrors.append(error)
    print('Final In-sample Error : ',errors[-1])
    print("Out of sample error : ",sum(testerrors)/numberOfTestSamples)
    
    compare_results[N] = errors
    

```

```python
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.title('Insample error v/s Epochs for sample size ')
plt.rcParams['figure.figsize'] = [10, 10]
for N in sample_lengths:
    plt.plot(range(len(compare_results[N])),compare_results[N],label='size_' + str(N))
plt.legend()
plt.show()
```
